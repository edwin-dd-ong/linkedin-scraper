```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import pandas as pd
import time


driver = webdriver.Chrome()

# Function to log in to LinkedIn
def linkedin_login(email, password):
    driver.get("https://www.linkedin.com/login")

    
    # Pause for manual security code input
    print("Please enter the security code manually.")
    input("Press Enter in this console once the code is entered and you're logged in...")

    

# Your LinkedIn credentials (consider using environment variables or input() for security)
your_email = "YOUR_EMAIL"
your_password = "YOUR_PASSWORD"

# Function to scrape profile information
def scrape_profile(url):
    driver.get(url)
    time.sleep(5)  # Wait for profile page to load
    data = {'linkedin_url': url}
    
    
    try:
        data['name'] = driver.find_element(By.CSS_SELECTOR, "h1").text
        experience_element = driver.find_element(By.XPATH, '//div[@id="experience"]/..')
        companies = get_organizations(experience_element, url)
        for i, company in enumerate(companies):
            data[f'company {i+1}'] = company
        education_element = driver.find_element(By.XPATH, '//div[@id="education"]/..')
        schools = get_organizations(education_element, url)
        for i, school in enumerate(schools):
            data[f'school {i+1}'] = school
    except Exception as e:
        print(f"Error extracting data for profile: {url} | Error: {e}")
    return data

def get_organizations(parent_element, url):
    try:
        parent_element.find_element(By.XPATH, './/a[contains(@id,"see-all") and @target="_self"]').click()
        time.sleep(5)  # Wait for page to load
        parent_element = driver.find_element(By.CLASS_NAME, "scaffold-layout__main")
        print("clicked on see all experiences/education")
    except Exception as e:
        # If it fails, just print the error and move on
        print(f"Could not click 'See All Experiences/Education'")
    logo_elements = parent_element.find_elements(By.XPATH, './/a[contains(@href,"company") and @target="_self"]')
    href_links = []
    for element in logo_elements:
        # Get the href attribute of each element and append it to the href_links list
        href_links.append(element.get_attribute("href"))
    href_links = list(set(href_links))
    organizations = []
    for link in href_links:
        driver.get(link)
        time.sleep(3)  # Wait for page to load
        try:
            # Attempt to find the h1 element
            h1_element = driver.find_element(By.CSS_SELECTOR, "h1")
            organizations.append(h1_element.text)
        except Exception as e:
            # If h1 element is not found, do nothing
            pass
    driver.get(url)
    time.sleep(5)
    return organizations
        

    
# List of LinkedIn profiles to scrape
profiles = [
    "https://linkedin.com/in/anderthan-hsieh-b0068551",
    "https://linkedin.com/in/maruaditya",
    "https://linkedin.com/in/robin-seitz-12029970",
    "https://linkedin.com/in/ashwin-nandakumar-menon",
    "https://linkedin.com/in/saifabuhashish",
    "https://linkedin.com/in/sacellarius",
    "https://linkedin.com/in/andrew-wang-0b5526109",
    "https://linkedin.com/in/lucas-switzer-0a3756103",
    "https://linkedin.com/in/dsert",
    "https://linkedin.com/in/ckore",
    "https://linkedin.com/in/huberttsen",
    "https://linkedin.com/in/wu-julia",
    "https://linkedin.com/in/karanbh",
    "https://linkedin.com/in/ellenfkh",
    "https://linkedin.com/in/anmolsood",
    "https://linkedin.com/in/matt-tengtrakool",
    "https://linkedin.com/in/varun-patel-091715156",
    "https://linkedin.com/in/tim-kral-512b5b3",
    "https://linkedin.com/in/jett-hollister",
    "https://linkedin.com/in/akhterov",
    "https://linkedin.com/in/colerubin",
    "https://linkedin.com/in/samagra-sharma-4476bb135",
    "https://linkedin.com/in/williamwang77",
    "https://linkedin.com/in/distributed",
    "https://linkedin.com/in/sourav-choraria-9b389776",
    "https://linkedin.com/in/rosenfieldmichael",
    "https://linkedin.com/in/jpalakapilly",
    "https://linkedin.com/in/briar-smith",
    "https://linkedin.com/in/adityajoshi101",
    "https://linkedin.com/in/nankigrewal",
    "https://linkedin.com/in/yjwxiao",
    "https://linkedin.com/in/nikki-a-shah",
    "https://linkedin.com/in/alex-choi",
    "https://linkedin.com/in/cyrusjia",
    "https://linkedin.com/in/diederik-vink",
    "https://linkedin.com/in/ralphma",
    "https://linkedin.com/in/soohoonchoi",
    "https://linkedin.com/in/ryan-ali",
    "https://linkedin.com/in/rachelsylee",
    "https://linkedin.com/in/ashutoshpurohit97"
]

# DataFrame to hold scraped data
df_columns = ['linkedin_url', 'name'] + [f'company {i}' for i in range(1, 11)] + [f'school {i}' for i in range(1, 11)] # Adjust 11 based on max expected companies
df = pd.DataFrame(columns=df_columns)

# Login to LinkedIn
linkedin_login(your_email, your_password)

# Scrape profiles and accumulate data
for profile in profiles:
    profile_data = scrape_profile(profile)
    profile_data_df = pd.DataFrame([profile_data]) 
    df = pd.concat([df, profile_data_df], ignore_index=True)

# Close the WebDriver
driver.quit()

# Drop columns with all NaN values to adjust for the actual number of companies


# Display the DataFrame
df
```
